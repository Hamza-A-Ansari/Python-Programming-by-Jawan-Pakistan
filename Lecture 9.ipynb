```python
def frq_table(data_body,col_num):
    dictionary = {}
    for i in data_body:
        value = i[col_num]
        if value not in dictionary:
            dictionary[value] = 1
        else:
            dictionary[value] += 1    
    return dictionary
```

```python
frq_table(data_body,7)
```
